```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
house_data = pd.read_csv("/users/amulya/Downloads/HousePricePrediction.csv")
filtered_data = house_data.dropna(subset=['SalePrice'])
saleprice_values = filtered_data['SalePrice'].values
# print(filtered_data)
filtered_data = filtered_data['SalePrice'].astype(int)
skewness = skew(filtered_data)
mean = filtered_data.mean()
median = filtered_data.median()
# print(max_whisker_value)
min_value = np.min(saleprice_values)
max_value = np.max(saleprice_values)
q1 = np.percentile(saleprice_values, 25)
q3 = np.percentile(saleprice_values, 75)
whisker_range = 1.5 * (q3 - q1)
max_whisker_value = q3 + whisker_range
print(f"{min_value}\n {max_value}\n (Q1): {q1}\n (Q3): {q3}")
plt.axvline(q1, color='gray', linestyle='--', label='Q1 Line')
plt.axvline(q3, color='orange', linestyle='--', label='Q3 Line')
plt.axvline(min_value, color='purple', linestyle='--', label='Min Line')
plt.axvline(max_whisker_value, color='brown', linestyle='--', label='Calculated Max Whisker Line')

plt.legend()
house_data['SalePrice'].plot(kind='box',vert=False)

```
