```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
house_data = pd.read_csv("/users/amulya/Downloads/HousePricePrediction.csv")

# Extract the 'LotArea' and 'SalePrice' columns
selected_columns = ['LotArea', 'SalePrice']
selected_data = house_data[selected_columns].dropna()

# Create a pair plot
sns.pairplot(selected_data, height=3, aspect=1.5)
plt.suptitle("Pair Plot of Lot Area and Sale Price", y=1.02)

# Show the plot
plt.show()
```
