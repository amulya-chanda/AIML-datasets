```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
loan_prediction_details = pd.read_csv("/users/amulya/Downloads/loan_approval_dataset.csv")
loan_prediction_details.columns = loan_prediction_details.columns.str.strip()
loan_prediction_details['loan_status'] = loan_prediction_details['loan_status'].replace({' Approved': 1, ' Rejected': 0})
corr_of_no_of_dependents = loan_prediction_details['no_of_dependents'].corr(loan_prediction_details['loan_status'])
corr_of_income_annum = loan_prediction_details['income_annum'].corr(loan_prediction_details['loan_status'])
corr_of_loan_amount = loan_prediction_details['loan_amount'].corr(loan_prediction_details['loan_status'])
corr_of_residential_assets_value = loan_prediction_details['residential_assets_value'].corr(loan_prediction_details['loan_status'])
corr_of_cibil_score = loan_prediction_details['cibil_score'].corr(loan_prediction_details['loan_status'])
corr_of_luxury_assets_value = loan_prediction_details['luxury_assets_value'].corr(loan_prediction_details['loan_status'])
corr_of_bank_asset_value = loan_prediction_details['bank_asset_value'].corr(loan_prediction_details['loan_status'])
corr_of_commercial_assets_value = loan_prediction_details['commercial_assets_value'].corr(loan_prediction_details['loan_status'])
# print(loan_prediction_details['loan_status'])
print(f'dependents : {corr_of_no_of_dependents} \n income : {corr_of_income_annum}\n loan : {corr_of_loan_amount}\n rs : {corr_of_residential_assets_value}\n cibil_score : {corr_of_cibil_score}\n luxury : {corr_of_luxury_assets_value}\n bank : {corr_of_bank_asset_value}\n commercial : {corr_of_commercial_assets_value}')
```

```python
income_annum = loan_prediction_details['income_annum']

# Define the bins for income ranges
bins = [0, 100000, 2000000, 3000000, 4000000, 5000000, 6000000, 7000000, 8000000, 9000000, 1000000000]  # Add more bins as needed

# Create labels for each bin
labels = ['0-1000000', '1000000-2000000', '200000-3000000', '3000000-4000000', '4000000-5000000', '5000000-6000000', '6000000-7000000', '7000000-8000000', '8000000-9000000', '9000000-100000000']

# Use pd.cut() within value_counts() to count the number of people in each income range
income_counts = pd.cut(income_annum, bins=bins, labels=labels, include_lowest=True).value_counts()
print(income_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
features = ['income_annum', 'loan_amount', 'cibil_score']
loan_status = loan_prediction_details['loan_status']

# Pairplot to visualize relationships
sns.pairplot(loan_prediction_details, vars=features, hue='loan_status', palette='husl')
plt.suptitle('Pairplot of Features by Loan Status', y=1.02)
plt.show()

# Box plots to compare distributions by loan status
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(15, 5))

for i, feature in enumerate(features):
    sns.boxplot(x='loan_status', y=feature, data=loan_prediction_details, ax=axes[i])
    axes[i].set_title(f'{feature} by Loan Status')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='loan_status', data=loan_prediction_details)
plt.title('Count Plot of loan status')
plt.xlabel('loan status')
plt.ylabel('Count')
plt.show()
```
