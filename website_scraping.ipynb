```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import TimeoutException, WebDriverException
from webdriver_manager.chrome import ChromeDriverManager
df['h1_text'] = ""  # Initialize all values to empty string
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
# sub_df = df['Website'].head(10)
# sub_df = pd.DataFrame(sub_df)
for index, row in df.iterrows():
    link = row['Website']
    try:
        driver.execute_script("window.open('');")
        driver.switch_to.window(driver.window_handles[-1])
        driver.get(link)
        
        # Find the first h1 element on the page
        elements = driver.find_elements(By.TAG_NAME, "h1")
        if elements:
            h1_text = elements[0].text  # Get text from the first h1 element found
            df.at[index, 'h1_text'] = h1_text
            print(f"Found h1 text on {link}: {h1_text}")
        else:
            df.at[index, 'h1_text'] = "No h1 found"
            print(f"No h1 text found on {link}")

        h2_elements = driver.find_elements(By.TAG_NAME, "h2")
        if h2_elements:
            h2_text = h2_elements[0].text  # Get text from the first h2 element found
            df.at[index, 'h2_text'] = h2_text
            print(f"Found h2 text on {link}: {h2_text}")
        else:
            df.at[index, 'h2_text'] = "No h2 found"
            print(f"No h2 text found on {link}")

        # Find the first p element on the page
        p_elements = driver.find_elements(By.TAG_NAME, "p")
        if p_elements:
            p_text = p_elements[0].text  # Get text from the first p element found
            df.at[index, 'p_text'] = p_text
            print(f"Found p text on {link}: {p_text}")
        else:
            df.at[index, 'p_text'] = "No p found"
            print(f"No p text found on {link}")
    
    except TimeoutException:
        print(f"Timeout while trying to load {link}")
        df.at[index, 'h1_text'] = "Timeout - Page could not be loaded"

    except WebDriverException as e:
        print(f"WebDriver exception occurred for {link}")
        df.at[index, 'h1_text'] = "Error loading page"

    finally:
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

driver.quit()

# Print or save the updated DataFrame
# print(df)
file_path = '/users/amulya/Desktop/csv_files/scrap_data.csv'
df.to_excel(file_path, index=False)  # Save the DataFrame to a CSV file
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import time
import pandas as pd

df = pd.read_excel('/users/amulya/Downloads/scraping_30.xlsx')
base_links = df['Website']

driver = webdriver.Chrome()
aboutPage = []
productPage = []
def check_url(url):
    try:
        response = requests.head(url, allow_redirects=True)
        if response.status_code == 200:
            return True
        return False
    except requests.RequestException:
        return False

def find_about_page(link):
    potential_urls = [
        f"{link}/about-us",
        f"{link}/about",
        f"{link}/aboutUS",
        f"{link}/About-Us",
        f"{link}/About",
        f"{link}/AboutUS"
    ]
    
    for url in potential_urls:
        if check_url(url):
            try:
                driver.get(url)
                time.sleep(2)  # Wait for the page to load
                body_elements = driver.find_elements(By.TAG_NAME, "body")
                aboutPage_data = "\n".join([element.text for element in body_elements])
                return aboutPage_data
            except Exception as e:
                print(f"Error accessing {url}: {str(e)}")
    return " "

def find_product_page(link):
    potential_urls = [
        f"{link}/products",
        f"{link}/product",
        f"{link}/Products",
        f"{link}/Product",
        f"{link}/products-overview",
        f"{link}/product-overview"
    ]
    
    for url in potential_urls:
        if check_url(url):
            try:
                driver.get(url)
                time.sleep(2)  # Wait for the page to load
                body_elements = driver.find_elements(By.TAG_NAME, "body")
                productPage_data = "\n".join([element.text for element in body_elements])
                return productPage_data
            except Exception as e:
                print(f"Error accessing {url}: {str(e)}")
    return " "

# Iterate through each base link
for link in base_links:
    try:
        aboutPage_data = find_about_page(link)
        aboutPage.append(aboutPage_data)
        productPage_data = find_product_page(link)
        productPage.append(productPage_data)
        print('#####################################################')
    except Exception as e:
        print(f"Error scraping data from {link}: {str(e)}")
        aboutPage.append(' ')
        productPage.append(' ')
        print('#######&&&&&&&&&&&&&&&&&&&&&&&&&&&&&')

home_page_data = []
# Iterate through each link
for link in base_links:
    try:
        # Open the link
        driver.get(link)
        
        # Wait for the page to load (you might need to adjust the time based on the page's loading speed)
        time.sleep(2)
        
        # Scraping data from the page
        # page_data = driver.find_elements(By.TAG_NAME, "body")
        # print("Data from", link, ":", page_data)
        body_elements = driver.find_elements(By.TAG_NAME, "body")
        page_data = "\n".join([element.text for element in body_elements])
        home_page_data.append(page_data)
        # df['Home Page'] = page_data
        print("Data from", link, ":", page_data)
        print('#####################################################')
    except Exception as e:
        print("Error scraping data from", link, ":", str(e))
        # df['Home Page'] = ' '
        home_page_data.append(' ')
        print('#####################################################')

# Close the web driver
driver.quit()
df['About Page'] = aboutPage
df['Home Page'] = home_page_data
df['Product Page'] = productPage
df.to_excel('/users/amulya/Desktop/webScraping_30.xlsx', index=False)
```
