```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
house_data = pd.read_csv("/users/amulya/Downloads/HousePricePrediction.csv")
print(house_data['SalePrice'])
# print(house_data.to_string())
filtered_data = house_data.dropna(subset=['SalePrice'])
saleprice_values = filtered_data['SalePrice'].values
mean_saleprice = filtered_data['SalePrice'].mean()
print(mean_saleprice)
median = filtered_data['SalePrice'].median() 
std_saleprice = filtered_data['SalePrice'].std(ddof = 0)
print(median)
x = np.linspace(mean_saleprice - 3*std_saleprice, mean_saleprice + 3*std_saleprice, 100)
y = (1/(std_saleprice * np.sqrt(2*np.pi))) * np.exp(-(x-mean_saleprice)**2 / (2*std_saleprice**2))
# Plot the histogram of SalePrice
plt.hist(saleprice_values, bins=30, density=True, alpha=0.6, color='g', label='SalePrice Histogram')

plt.xlabel('values')
plt.ylabel('Saleprice') 

plt.axvline(mean_saleprice, color='gray', linestyle='--', label='mean Line')
plt.axvline(median, color = 'yellow', linestyle='--', label='meadian Line')
plt.legend()
# Show the plot
plt.show()

house_data['SalePrice'].plot(kind='box',vert=False)
sale_price_df = pd.DataFrame(house_data['SalePrice'])
```
